# Sistema de ecuaciones diferenciales a resolver 
$$
\dot{x} = 10(y-z)
$$
$$
\dot{y} = x(28-z) - y 
$$
$$
\dot{z} = xy - \frac{8}{3}z
$$
con las condiciones iniciales aleatorias:
$$
x(0),y(0),z(0) ∈ [-5,5] 
$$

```python
def f1(x,y,z): 
    return 10*(y-z)
def f2(x,y,z): 
    return x*(28-z) -y
def f3(x,y,z): 
    return x*y -(8/3)*z
```

# Metodo de Euler

```python
def euler(funcion1,funcion2,funcion3,p_ini,t_final,h): 
    tiempos = arange(0,t_final+h,h) 
    x = [0]*len(tiempos) 
    y = [0]*len(tiempos) 
    z = [0]*len(tiempos) 
    x[0] = p_ini[0]  
    y[0] = p_ini[1] 
    z[0] = p_ini[2]
    for i in range(len(tiempos)-1): 
        x[i+1] =  x[i] + funcion1(x[i],y[i],z[i])*h 
        y[i+1] =  y[i] + funcion2(x[i],y[i],z[i])*h 
        z[i+1] =  z[i] + funcion3(x[i],y[i],z[i])*h 
    return tiempos,x,y,z
```

Se elije una condición inicial aleatoria en el x, y, z ∈ [−5, 5]

```python
import random
p = random.randrange(-5,5),random.randrange(-5,5),random.randrange(-5,5)
print "condicion inicial aleatoria (x(0),y(0),z(0)) =", p
sol = euler(f1,f2,f3,[p[0],p[1],p[2]],1.,.0001)# tiempo final = 1 y h= .0001
```

# Plano XY

```python
plot(sol[1],sol[2])
show()
```

# Plano XZ

```python
plot(sol[1],sol[3])
show()
```

# Plano YZ

```python
plot(sol[2],sol[3])
show()
```

# Se observa una espiral en 3 dimensiones, semejante a un resorte

Ahora se toman 3 soluciones parecidas. Para ello se proposieron 3 condiciones iniciales tales que en la coordenada en y las condiciones iniciales difieran entre si en $10^{-2}$

```python
sol1 = euler(f1,f2,f3,[1,1,1],2.,.001) #(x(0),y(0),z(0)) = (1, 1, 1)
sol2 = euler(f1,f2,f3,[1,1+10**(-2),1],2.,.001) #(x(0),y(0),z(0)) = (1, 1 + .01, 1)
sol3 = euler(f1,f2,f3,[1,1+20**(-2),1],2.,.001) #(x(0),y(0),z(0)) = (1, 1 + .02, 1)
```

# Plano XY

```python
plot(sol1[1],sol1[2],sol2[1],sol2[2],sol3[1],sol3[2])
xlim([-5000,1500]),ylim([-150000,50000])
show()
```

# Plano YZ

```python
plot(sol1[2],sol1[3],sol2[2],sol2[3],sol3[2],sol3[3])
xlim([-7000,30000]),ylim([-22000,10000])
show()
```

# Plano XZ

```python
plot(sol1[1],sol1[3],sol2[1],sol2[3],sol3[1],sol3[3])
xlim([-800,1300]),ylim([-22000,10000])
show()
```

# Se realizaron integraciones con tiempos relativamente largos (t final = 2) en los que se pudieron apreciar diferencias en el comportamiento de las soluciones. El comportamiento de las soluciones en puntos cercanos a t = 0 es igual, mientras que en puntos muy alejados se aprecian discrepancias entre las soluciones. Ya que el comportamiento de las soluciones solo cambia en puntos extremadamente alejados, para fines prácticos  las soluciones son iguales si se tienen pequeñas variaciones en una de las condiciones iniciales
